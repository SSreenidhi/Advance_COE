```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
import torch

model_name="t5-base"
tokenizer=T5Tokenizer.from_pretrained(model_name)
model=T5ForConditionalGeneration.from_pretrained(model_name)
```

```python
#step2 Input text for summarization
text="""The Taj Mahal, located in Agra, India, is one of the most iconic monuments in the world. Built by Mughal emperor Shah Jahan in memory of his beloved wife Mumtaz Mahal, it stands as a symbol of eternal love and devotion. Made of pure white marble and adorned with intricate carvings and precious stones, the Taj Mahal attracts millions of visitors every year for its breathtaking beauty and historical significance."""
```

```python
#step3: prepared task prefix "summarization"
input_text="summarize:"+text
```

```python
#Tokenize input
input_ids=tokenizer.encode(input_text,return_tensors="pt",max_length=512,truncation=True)
```

```python
#step 5: Generate summary
summary_ids=model.generate(input_ids,max_length=150,min_length=30,length_penalty=2.0,num_beams=4,early_stopping=True)
```

```python
summary_ids
```

```python
summary=tokenizer.decode(summary_ids[0],skip_special_tokens=True)
```
