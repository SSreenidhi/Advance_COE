```python
from transformers import PegasusTokenizer,PegasusForConditionalGeneration
import torch

model_name = 'google/pegasus-xsum' # Corrected model name
tokenizer=PegasusTokenizer.from_pretrained(model_name)
model=PegasusForConditionalGeneration.from_pretrained(model_name)
```

```python
# Example text to summarize
text = """
The Pegasus-XSum model is a large language model fine-tuned for abstractive summarization of news articles.
Abstractive summarization aims to generate concise, fluent summaries that may include words and phrases not present in the original document.
This differs from extractive summarization, which only selects and rearranges sentences from the source text.
Pegasus-XSum is particularly good at generating short, single-sentence summaries.
"""
```

```python
# Tokenize the input text
input_ids = tokenizer(text, return_tensors="pt").input_ids
```

```python
# Generate the summary
output_ids = model.generate(input_ids, max_length=32, num_beams=5, early_stopping=True)
```

```python
# Decode the generated summary
summary = tokenizer.decode(output_ids[0], skip_special_tokens=True)
```

```python
print("Original Text:")
print(text)
print("\nSummary:")
print(summary)
```
