```python
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)
```

# **Model1**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model = Sequential()

# CONV_1: 2D conv layer with RELU activation and depth = 32 kernels
model.add(Conv2D(32, kernel_size=(3, 3), padding='same', activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

# CONV_2: 2D conv layer with RELU activation and depth = 64 kernels
model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(10, activation='softmax'))

model.summary()
```

```python
X_train[0].ndim
```

```python
model.compile(loss='categorical_crossentropy',optimizer='rmsprop',metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.callbacks import EarlyStopping

early_stop=EarlyStopping(monitor='val_loss',
                            patience=4,
                            restore_best_weights=True,
                         verbose=1)
checkpointer=ModelCheckpoint(filepath='model1.keras',
                             verbose=1,
                             save_best_only=True)
```

```python
hist=model.fit(X_train,y_train,batch_size=80,epochs=10,
               validation_data=(X_test,y_test),
               callbacks=[early_stop,checkpointer],
               verbose=2,shuffle=True)
```

# **Model2**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model2 = Sequential()

# CONV_1: 2D conv layer with RELU activation and depth = 32 kernels
model2.add(Conv2D(30, kernel_size=(2, 2), padding='same', activation='relu', input_shape=(28, 28, 1)))
model2.add(MaxPooling2D(pool_size=(2, 2)))
model2.add(Dropout(0.25))

# CONV_2: 2D conv layer with RELU activation and depth = 64 kernels
model2.add(Conv2D(64, (2, 2), padding='same', activation='relu'))
model2.add(MaxPooling2D(pool_size=(2, 2)))

model2.add(Flatten())
model2.add(Dense(60, activation='relu'))
model2.add(Dropout(0.3))
model2.add(Dense(16, activation='relu'))
model2.add(Dense(10, activation='softmax'))

model2.summary()
```

```python
model2.compile(loss='categorical_crossentropy',optimizer='rmsprop',metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.callbacks import EarlyStopping

early_stopp=EarlyStopping(monitor='val_loss',
                            patience=4,
                            restore_best_weights=True,
                         verbose=1)
checkpoint=ModelCheckpoint(filepath='model1.keras',
                             verbose=1,
                             save_best_only=True)
```

```python
hist2=model2.fit(X_train,y_train,batch_size=80,epochs=20,
               validation_data=(X_test,y_test),
               callbacks=[early_stopp,checkpoint],
               verbose=2,shuffle=True)
```
