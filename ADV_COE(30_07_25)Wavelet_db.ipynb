```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
embedding_model_name='sentence-transformers/all-MiniLM-L6-v2'
```

```python
embedding=HuggingFaceEmbeddings(
    model_name=embedding_model_name)
```

```python
from langchain.document_loaders import PyPDFLoader
```

```python
path="/content/drive/MyDrive/Cast iron.pdf"
loader=PyPDFLoader(path)
pages=loader.load()
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=300,chunk_overlap=20)
docs=text_splitter.split_documents(pages)
```

```python
len(docs)
```

```python
from langchain_community.vectorstores import Weaviate
vector_db=Weaviate.from_documents(
    docs,embedding,client=client,by_text=False
)
```

```python
print(vector_db.similarity_search("what is the cast iron?",k=3)[0].page_content)
```

```python
from langchain.prompts import PromptTemplate
```

```python
template=(
    "You are assistant to question-answering tasks. \n"
    "Use the following pieces of retrived context to answer the question. \n"
    "If you don't know the answer, just say that you don't know, don't try to make up an answer. \n"
    "Use three sentences maximum and keep the answer concise. \n"
    "Question: {question}\n"
    "Context:{context}\n"
    "Helpful Answer:"
)
prompt=prompt=PromptTemplate.from_template(template)
```

```python
from google.colab import userdata
hf_api_token=userdata.get('hf_api_token')
```

```python
model_id='gpt2'
```

```python
from transformers import AutoTokenizer,AutoModelForCausalLM,pipelines
tokenizer=AutoTokenizer.from_pretrained(model_id)
model=AutoModelForCausalLM.from_pretrained(model_id)

generator=pipelines.pipeline(
    'text-generation',
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=300,
    temperature=0.3
)
```

```python
from langchain_community.llms import HuggingFacePipeline
llm2=HuggingFacePipeline(pipeline=generator)
```

```python
from langchain.schema.runnable import RunnablePassthrough
from langchain.schema.output_parser import StrOutputParser
```

```python
output_parser=StrOutputParser()
receiver=vector_db.as_retriever()
```

```python
rag_chain=(
    {'context':receiver, 'question':RunnablePassthrough()}
    |prompt
    |llm2
    |output_parser
)
```

```python
client=weaviate.Client(
    url=WEAVIATE_URL,
    auth_client_secret=AuthApiKey(api_key=WEAVIATE_API_KEY)
)
```

```python
response=rag_chain.invoke('What is cast iron?')
```
